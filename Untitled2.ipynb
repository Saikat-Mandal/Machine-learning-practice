```python
df = pd.read_csv("90 - heart-disease.csv")
df.head()
```

```python
over = df[df["age"]>50]
len(over)
```

```python
over.plot( x="age" , y ="chol" ,kind="scatter" , c="target")
```

```python
# oo method with pyplot method mixed
fig , ax = plt.subplots(figsize=(10,6))
over.plot(kind ="scatter",
         x="age",
         y="chol",
         c="target",
         ax=ax);
```

```python
# oop method from scratch 
fig , ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over["age"] , y=over["chol"] , c=over["target"])
ax.set(title="heart disease" , xlabel="age" , ylabel="cholestrol")

# add a legend
ax.legend(*scatter.legend_elements() , title="target");

# add horizontal line
ax.axhline( over['chol'].mean() ,linestyle="--")
```

```python
fig , (ax0 , ax1) = plt.subplots(nrows=2 , ncols=1 , figsize=(10,10))
# add data to ax)
scatter = ax0.scatter(x=over["age"] , y = over["chol"] , c=over["target"]);
# customize ax0 
ax0.set(title="over" , xlabel="age" , ylabel="cholestrol");
ax0.legend(*scatter.legend_elements() , title="target");
##########################################################################
# add data to ax1 
scatter2 = ax1.scatter(x=over["age"] , y =over["thalach"] , c=over["target"])
ax1.set(title="over2" , xlabel="thalach" , ylabel="age")
ax1.legend(*scatter2.legend_elements() , title="target")

# add title to the figure 
fig.suptitle("heart title")
```
